```python
import cv2
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15,15)

img = cv2.imread("wallpaper.jpg",0) # Read as gray scale image
plt.imshow(img, cmap='gray') #since original color map is purple-ish
```

```python
ret, thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)

plt.imshow(thresh1, cmap='gray')
```

```python
ret, thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)

plt.imshow(thresh1, cmap='gray')
```

```python
ret, thresh1 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)

plt.imshow(thresh1, cmap='gray')
```

```python
ret, thresh1 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)

plt.imshow(thresh1, cmap='gray')
```

```python
new_image = cv2.imread("crossword.png",0)

plt.figure(figsize=(10,10))
plt.imshow(new_image,cmap='gray')
```

```python
ret, th1 = cv2.threshold(new_image, 200, 255, cv2.THRESH_BINARY)
print(ret)
plt.imshow(th1, cmap='gray')
```

```python
th2 = cv2.adaptiveThreshold(new_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 
                                11,8)

plt.imshow(th2, cmap='gray')
```

```python
blended = cv2.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4,gamma=0)

plt.imshow(blended, cmap='gray')
```
