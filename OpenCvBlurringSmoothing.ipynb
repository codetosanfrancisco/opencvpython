```python
# Blurring is often combined with Edge Detection
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    image = cv2.imread("wallpaper.jpg").astype(np.float32) / 255
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
def show_img(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111) # 1 row, 1 column, First place
    ax.imshow(img)
```

```python
show_img(load_img())
```

```python
gamma = 8
res = np.power(load_img(), gamma)
show_img(res)
```

```python
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX

cv2.putText(img, text="Mountain", org=(1000,2000), fontFace=font, fontScale=20, color=(255,0,0), thickness=5)

show_img(img)
```

```python
kernel = np.ones(shape=(5,5), dtype=np.float32) / 25
kernel
```

```python
result = cv2.filter2D(img,-1,kernel)
```

```python
show_img(result)
```
